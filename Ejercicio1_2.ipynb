## Ejercicio 2 
Dado un archivo que contiene en cada línea una palabra o conjunto de palabras seguido de un valor numérico denominado “sentimiento” y un conjunto de tweets, se pide calcular el sentimiento de aquellas palabras o conjunto de palabras que no tienen un valor asociado en el archivo de “sentimientos”. Se pueden seguir distintas estrategias para asignar un valor. Por ejemplo, se podría asignar como valor el valor del “sentimiento” del tweet en que se encuentra la palabra o conjunto de palabras sin valor, o el valor medio del “sentimiento” del tweet.

## Solucion
La solucion propuesta consta de cuatro funciones:

·def cargar_sentimientos(Sentimientos): Cuya entrada es el nombre del fichero que contiene los sentimientos y cuya salida es un diccionario cuyas claves son los nombres de los sentimientos y cuyos valores son los valores enteros asociados a esos sentimientos (esta funcion se proporciono como parte del enunciado)

·def procesar_json(Tweets): Cuya entrada es el nombre del fichero que contiene los tweets, y cuya salida es un array de diccionarios, uno por cada tweet que contiene el fichero (la longitud del array es el numero de tweets).

·def ejercicio_2(Sentimientos, Tweets): Esta es la funcion principal del ejercicio y llama a las dos funciones anteriores. Primero carga el diccionario de sentimientos, luego carga el array de tweets, y  recorre el array de tweets para analizar el texto de cada uno y calcular el valor sentimental. Una vez tiene el valor sentimental de cada tweet, 
devuelve un nuevo diccionario que contiene todas las palabras sin valor en el diccionario de sentimientos, con 
su valor calculado como la mediana de los valores de los tweets en los que aparece.

·def imprimir_por_pantalla(dic, ordenar): Es una funcion que dado un diccionario de palabra y valor sentimental, muestra por pantalla cada palabra (clave del diccionario) y el valor asignado a cada una. Ofrece la posibilidad de ordenarlas alfabeticamente, indicando como parametro ordenar=True.Por defecto, el parametro ordenar esta a False.

El programa pide que se introduzcan por teclado los dos archivos (Sentimientos y Tweets) y si se quiere la salida por orden alfabetico. Luego llama a la funcion ejercicio_2 para generar el diccionario y a imprimir_por_pantalla para mostrar la informacion solicitada de la siguiente forma:

Los valores sentimentales de las palabras son:

i	1
whale	0
slap	0
your	-1
fin	0
and	2



```python
#importar modulos
import json
import numpy as np

#Funcion que lee el archivo de sentimientos y devuelve un diccionario cargado con los calores int de los sentimientos
def cargar_sentimientos(file1):
    sentimientos=open(file1)
    valores={}
    for linea in sentimientos:
        termino,valor=linea.split("\t")
        valores[termino]= int(valor)
        
    return valores


#Procesado del fichero json que devuelve un array de diccionarios (uno por tweet)
def procesar_json(file):
    paraprocesar=open(file,'r')
    cadena=paraprocesar.read()
    paraprocesar.close()

    replaced=cadena.replace("}\n{", "},{") #reemplazando "}/n{" por "}{"
    replaced='['+replaced+']'              #abriendo y cerrando con corchetes para convertir la cadena en un array de diccionarios
    
    output= open('Tweets.json','w')
    output.write(replaced)
    output.close()
    #Carga de los tweets del fichero
    with open('Tweets.json','r')as file:
        try: tweets=json.load(file)
        except:  tweets=None
    #  Carga finalizada
    return tweets #Devuelve un array de diccionarios



#Funcion ppal ejercicio 2: Como parametros de entrada tiene el archivo de sentimientos y el archivo de tweets,
#y  devuelve un nuevo diccionario que contiene todas las palabras sin valor en el diccionario de sentimientos, con 
#su valor calculado como la mediana de los valores de los tweets en los que aparece.

def ejercicio_2(file_sent, file_tweets): 
    
    #Inicializacion de variables
    diccionario_sentimental=cargar_sentimientos(file_sent)
    k="text"
    sent_claves=diccionario_sentimental.keys()
    tweets=procesar_json(file_tweets)
    dic_palabras_sin_valor={}

    for x in range(0,len(tweets)):  #recorriendo el array de tweets
        if k in tweets[x].keys(): #si el tweet tiene texto
            texto=tweets[x].get(k).lower()#asigno el texto del tweet a una cadena, convirtiendolo todo a minusculas
            palabras=texto.split()
            #Para cada palabra del tweet, miro si esta en el diccionario de sentimientos, y sumo su valor
            valor_tweet=0
            for sentimiento in palabras: 
                if sentimiento in sent_claves:
                    valor_sentimiento=int(diccionario_sentimental.get(sentimiento))
                    valor_tweet+=valor_sentimiento
                else: pass


            # Separo la cadena en palabras, y las que no esten el el diccionario de sentimientos, las añado a un nuevo diccionario, 
            # y les asigno el valor sentimental del tweet. Si ya han aparecido antes, adjunto el nuevo valor para al final hacer la media
            # media

            for palabra in palabras:  #para cada palabra del tweet
                if palabra.isalpha(): #que es alfabetica (para no contar los emoticonos o los signos de puntuacion o las @)
                    if palabra not in sent_claves: #y no tiene valor en el archivo de sentimientos
                        if palabra in dic_palabras_sin_valor.keys(): # si la he adjuntado antes, le añado el valor del nuevo tweet para hacer luego la media
                            dic_palabras_sin_valor[palabra]+= (","+str(valor_tweet))
                        else:
                            dic_palabras_sin_valor[palabra]=str(valor_tweet) #si no estaba en el diccionario la añado
                    else: 
                        pass
                else:
                    pass
        else: pass  #cuando el tweet no tiene texto no hago nada   

    for palabra in dic_palabras_sin_valor.keys(): #Para cada palabra calculo la media y la reasigno al valor
        cadena=dic_palabras_sin_valor[palabra]
        lista=list(map(int, (cadena.split(","))))#convierto los valores a enteros
        valor=int(round(np.mean(lista)))                     #calculo la media y la convierto a entero
        dic_palabras_sin_valor[palabra]=valor    #se la asigno al valor en el diccionario
    return dic_palabras_sin_valor

##Funcion para imprimir por pantalla el diccionario, se le puede indicar si lo quieres ordenar o no
def imprimir_por_pantalla(dic,ordenar=False):
    if ordenar:
        for palabra in sorted(dic.keys()):
            print(palabra+"\t"+str(dic[palabra]))
    else:        
        for palabra in dic.keys():
            print(palabra+"\t"+str(dic[palabra]))
            

#Peticion de parametros de entrada y llamada al ejercicio:

print("Por favor, indique el fichero que contiene el valor sentimental de cada palabra")
file_dic=input()
print("Por favor, indique el fichero que contiene los tweets")
file_tweets=input()
print("¿Quiere mostrar los resultados por orden alfabético? (Y/N)")
ordenar=input()
dic=ejercicio_2(file_dic,file_tweets)
print("\nLos valores sentimentales de las palabras son:\n")

imprimir_por_pantalla(dic,ordenar.lower()=="y")

    



```
