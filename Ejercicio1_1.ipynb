# Ejercicio 1

Dado un fichero (Tweets) que contiene una serie de tweets, y otro fichero (Sentimientos) que contiene un diccionario de sentimientos y un valor asociado a cada uno, se trata de asociar a cada tweet del fichero Tweets un valor sentimental, que sera la suma de los valores de los sentimientos que contenga.

## Solucion
La solucion propuesta consta de tres funciones:

Â·def cargar_sentimientos(Sentimientos):
Cuya entrada es el nombre del fichero que contiene los sentimientos y cuya salida es un diccionario cuyas claves son los nombres de los sentimientos y cuyos valores son los valores enteros asociados a esos sentimientos (esta funcion se proporciono como parte del enunciado)

Â·def procesar_json(Tweets):
Cuya entrada es el nombre del fichero que contiene los tweets, y cuya salida es un array de diccionarios, uno por cada tweet que contiene el fichero (la longitud del array es el numero de tweets).

Â·def ejercicio_1(Sentimientos, Tweets):
Esta es la funcion principal del ejercicio y llama a las dos funciones anteriores.
Primero carga el diccionario de sentimientos, lueo carga el array de tweets, y por ultimo, recorre el array de tweets para analizar el texto de cada uno, calcular el valor sentimental y mostrarlo por pantalla.


El programa pide que se introduzcan por teclado los dos archivos (Sentimientos y Tweets) y llama a la funcion ejercicio_1 para mostrar por pantalla la informacion solicitada de la siguiente forma:

El siguiente tweet, con id "Tweet_id" "Tweet_text"\n
Tiene un sentimiento de: "Tweet_value"



```python
#importar modulos
import json

#Funcion que lee el archivo de sentimientos y devuelve un diccionario cargado con los calores int de los sentimientos
def cargar_sentimientos(file1):
    sentimientos=open(file1)
    valores={}
    for linea in sentimientos:
        termino,valor=linea.split("\t")
        valores[termino]= int(valor)
        
    return valores


#Procesado del fichero json que devuelve un array de diccionarios (uno por tweet)
def procesar_json(file):
    paraprocesar=open(file,'r')
    cadena=paraprocesar.read()
    paraprocesar.close()

    replaced=cadena.replace("}\n{", "},{") #reemplazando "}/n{" por "}{"
    replaced='['+replaced+']'              #abriendo y cerrando con corchetes para convertir la cadena en un array de diccionarios
    
    output= open('Tweets.json','w')
    output.write(replaced)
    output.close()
    #Carga de los tweets del fichero
    with open('Tweets.json','r')as file:
        try: tweets=json.load(file)
        except:  tweets=None
    #  Carga finalizada
    return tweets #Devuelve un array de diccionarios

#Funcion ppal ejercicio 1: Como parametros de entrada tiene el archivo de sentimientos y el archivo de tweets,
#y como resultado imprime por pantalla el  valor sentimental calculado.
def ejercicio_1(file_sent, file_tweets): 
    #Carga del diccionario de entimientos
    diccionario_sentimental=cargar_sentimientos(file_sent)
    k="text"
    sent_claves=diccionario_sentimental.keys()
    
    #Carga de los tweets del fichero
    tweets=procesar_json(file_tweets)
    
    #Procesado de los tweets uno a uno
   
    for x in range(0,len(tweets)):  #recorriendo el array de tweets
        if k in tweets[x].keys(): #si el tweet tiene texto
            texto=tweets[x].get(k).lower()#asigno el texto del tweet a una cadena
            palabras=texto.split()
            #Para cada palabra del tweet, miro si esta en el diccionario de sentimientos, y sumo su valor
            valor_tweet=0
            for sentimiento in palabras: 
                if sentimiento in sent_claves:
                    valor_sentimiento=int(diccionario_sentimental.get(sentimiento))
                    valor_tweet+=valor_sentimiento
                else: pass
            print("El siguiente tweet, con id "+ str(tweets[x].get('id'))+' "'+texto+'"')
            print("Tiene un sentimiento de: " + str(valor_tweet)+"\n")
    
        else: pass  
        

#Peticion de parametros de entrada y llamada al ejercicio:

print("Por favor, indique el fichero que contiene el valor sentimental de cada palabra")
file_dic=input()
print("Por favor, indique el fichero que contiene los tweets")
file_tweets=input()
print("\n")
ejercicio_1(file_dic,file_tweets)
```
